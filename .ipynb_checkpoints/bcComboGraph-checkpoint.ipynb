```python
import pickle
from collections import Counter
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms.mis import maximal_independent_set
import numpy as np
import time, os
from itertools import combinations, product
from copy import deepcopy
import pandas as pd, numpy as np
from itertools import product
```

# 3-on 5-off barcode

```python
totalCycles = 8
offCycles = 5
uniColorAllowed = False
prefix = "240508_3on-5off-3col_"
panel_gene_list_file = '~/workspaces/dartfish/ProbeDesign/Probe_design_MCrossTissue_3on_5off_0824/Mouse Cross Tissue Panel 09052024.csv'
#Generate Barcode List - each row a barcode
barcodelist = np.array(list(product([0, 1, 2, 3], repeat = totalCycles)))

# enforce the number of off-cycles
barcodelist = barcodelist[(barcodelist == 0).sum(axis = 1) == offCycles, ]

# enforce the multi-color barcode
barcodelist = pd.DataFrame(barcodelist)
if not uniColorAllowed:
    if offCycles > 0:
        barcodelist = barcodelist.loc[barcodelist.nunique(axis = 1) >= 3]
    else:
        barcodelist = barcodelist.loc[barcodelist.nunique(axis = 1) >= 2]

# shuffling the rows of the barcode list
barcodelist = barcodelist.sample(frac = 1).reset_index(drop = True)
# barcodelist.to_csv(f"./barcodes/{prefix}all.tsv", sep="\t")
barcodelist
```

```python
# finding all barcodes with hamming distance >=3 
bclist = barcodelist.copy()
def hamming(a, b):
    return len(a) - (a==b).sum()

distthresh1 = 3

inds = [] # indices of entries with mutually high Hamming distance

for i in bclist.index:
    isbad = False
    
    for j in inds:

        if hamming(bclist.loc[j], bclist.loc[i]) < distthresh1:
            isbad = True
            break

    if not isbad:
        inds.append(i)
    if i % 200 == 0:
        print(i)

q_df = bclist.loc[inds]
os.makedirs('./barcodes/',exist_ok=True)
q_df.to_csv(f"./barcodes/{prefix}Hamming3shuffle.tsv", sep="\t")
q_df.shape
```

```python
# creating a graph with pairs of barcodes connected only if they hamming distance == 1
allInds = barcodelist.index.values

G = nx.Graph()
G.add_nodes_from(allInds)

cnt = 0
# t1 = time()
for ind1, ind2 in combinations(barcodelist.index.values, 2):
    if hamming(barcodelist.loc[ind1], barcodelist.loc[ind2]) == 1:
        cnt += 1
        G.add_edge(ind1, ind2)

# Finding maximal independent set which gives us a set of barcodes in which no pair are adjacent (hamming > 1)
runs, lengths = [], [] 
for i in range(20000):
    runs.append(maximal_independent_set(G, nodes= list(q_df.index.values)))
    lengths.append(len(runs[-1]))
print(np.max(lengths))
```

```python
# taking the best run
best_h2_inds = sorted(runs[np.argmax(lengths)])
best_h2 = barcodelist.loc[best_h2_inds]
best_h2.to_csv(f"./barcodes/{prefix}Hamming2_network.tsv", sep="\t")

onehot_dict = {0 : '000', 1 : '001', 2 : '010', 3 : '100'}
best_h2_bin = best_h2.applymap(lambda x: onehot_dict[x])
best_h2_bin['joined'] = best_h2_bin.values.sum(axis=1)
best_h2_bin.to_csv(f"./barcodes/{prefix}Hamming2_network_binary.tsv", sep="\t")
best_h2_bin
```

```python
# Visualize selected barcodes within all barcodes
import matplotlib.pyplot as plt
# H = G.subgraph(q_df.index.values)
plt.figure(figsize=(15,15))
color = ['r' if i in best_h2_inds else 'b' for i in list(range(len(barcodelist)))]
pos = nx.spring_layout(G, k=0.1)
nx.draw_networkx(G,pos,  with_labels=False, font_weight='bold', node_color = color, alpha=0.6, node_size=20)
plt.savefig(f"./barcodes/{prefix}Graph of selected barcodes within barcodes.pdf")
```

```python
# Visualize to confirm there isn't any edges within selected barcodes
import matplotlib.pyplot as plt
G_selected = G.subgraph(best_h2_inds)
G_unselected = G.subgraph([x for x in barcodelist.index.to_list() if x not in best_h2_inds])
fig, ax = plt.subplots(nrows=1, ncols=2,figsize = (12,6))
nx.draw_networkx(G_selected,  with_labels=False, font_weight='bold', node_color = 'r', alpha=0.6, node_size=20,ax=ax[0])
ax[0].set_title('Selected barcodes')
pos = nx.spring_layout(G, k=0.12)
nx.draw_networkx(G_unselected, pos, with_labels=False, font_weight='bold', node_color = 'b', alpha=0.6, node_size=20,ax=ax[1])
ax[1].set_title('Unselected barcodes')
plt.savefig(f"./barcodes/{prefix}Separate graph for selected or unselected barcodes.pdf")
```
